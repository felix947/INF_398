```python
X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
scaler=MinMaxScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
np.random.seed(1)
model=Ridge()
x=np.linspace(0,1,25)
param_grid={'alpha':x,'random_state':[1]}
ridge_cv=GridSearchCV(model,param_grid,verbose=2,scoring='neg_root_mean_squared_error',cv=2)
ridge_cv.fit(X_train,y_train)
```

```python
ridge_cv.best_params_
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,ridge_cv.best_estimator_.predict(X_test)))
```

```python
np.sqrt(mean_squared_error(y_train,ridge_cv.best_estimator_.predict(X_train)))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
param_grid={'max_features':['auto', 'sqrt', 'log2'],'splitter':['best'],'max_depth':[i for i in range(1,10)]}
model2=DecisionTreeRegressor()
tree_cv=GridSearchCV(model2,param_grid,verbose=2,scoring='neg_root_mean_squared_error',cv=5)
tree_cv.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,tree_cv.best_estimator_.predict(X_test)))
```
